## Project description and content of the dataset

This project is an attempt 

i) to find regression models that can best predict the metric burnout score with the inclusion of metric and categorical parameters given in the data set, and 

ii) to estimate the impact of various factors (work circumstances, personal factors) on possible burnout situations of company employees.

The dataset has been taken from https://www.kaggle.com/blurredmachine/are-your-employees-burning-out. <br>

Only the train dataset has been used, as the related test dataset doesn't contain values for the dependent variable which I would like to have, in order to evaluate performance of my models. The train dataset was downloaded and worked on locally.

The dataset consists of 9 columns as follows:

- Employee ID: The unique ID allocated for each employee (example: fffe390032003000)
- Date of Joining: The date-time when the employee has joined the organization (example: 2008-12-30)
- Gender: The gender of the employee (Male/Female)
- Company Type: The type of company where the employee is working (Service/Product)
- WFH Setup Available: Is the work from home facility available for the employee (Yes/No)
- Designation: The designation of the employee of work in the organization, in the range of [0.0, 5.0]. The bigger the number, the higher is designation.
- Resource Allocation: The amount of resource allocated to the employee to work, ie. number of working hours, in the range of [1.0, 10.0]. A higher number means more workload.
- Mental Fatigue Score: The level of fatigue mentally the employee is facing, in the range of [0.0, 10.0] where 0.0 means no fatigue and 10.0 means completely fatigue.
- Burn Rate: The value to be predicted for each employee, telling the rate of Burn out while working, in the range of [0.0, 1.0] where the higher the value is the higher is the burn out.

### Data preparation for machine learning

#### Imputing missing values

As discussed above, missing values in 'Ressource' and 'Fatigue' will be replaced by means calculated separately for each group built by intersection of 'Gender' and 'WFH'.

```python
for i in [train_set, test_set]:
    for j in ['Ressource', 'Fatigue']:
        i[j] = i[j].fillna(i.groupby(['Gender', 'WFH'])[j].transform('mean'))
```

#### Scaling metric variables using MinMaxScaler

```python
# MinMaxScaler scales and translates each feature individually 
# such that it is in the given range on the training set, e.g. between zero and one.

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
metric = ['Tenure', 'Designation', 'Ressource', 'Fatigue']

for i in metric:
    train_set[[i]] = scaler.fit_transform(train_set[[i]])
    test_set[[i]] = scaler.transform(test_set[[i]])
```

#### Converting categorial variables with OrdinalEncoder

'Gender', 'Company_type', and 'WFH' - there are only two values in each of these variables, so OrdinalEncoder is enough: it will convert the labels into 0 and 1.

```python
from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder()
```

```python
for i in [train_set, test_set]:
    for j in [['Gender', 'Company_type', 'WFH']]:
        i[j] = ordinal_encoder.fit_transform(i[j])
```

#### Separating X from y

```python
y_train = train_set.iloc[:, 7]
X_train = train_set.iloc[:, 0:-1]
y_test = test_set.iloc[:, 7]
X_test = test_set.iloc[:, 0:-1]
```

```python
# Converting the datasets into numpy arrays
y_train = y_train.to_numpy()
X_train = X_train.to_numpy()
y_test = y_test.to_numpy()
X_test = X_test.to_numpy()
```

## Machine learning part

As the response variable is metric, I select four regressors which should be able to deliver metric predictions:  Linear Regression, Decision Tree Regressor, Random Forest Regressor, and Support Vector Regressor.

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
random_state=1
```

The following ratios will be used for the evaluation and comparison of the models:

- Root-mean-square error - indicates how much a point estimator scatters around the value to be estimated.
- R2 - is used to assess the goodness of fit of a regression and describes the share of the sum of squares explained by the regression in the total sum of squares to be explained. For my models, I will try to target R2=0.9 as the minimum acceptable value.

### Initial fitting

Four selected regressors will be first applied with their basic parameters.

```python
lin_reg = LinearRegression()
tree_reg = DecisionTreeRegressor(random_state=random_state)
forest_reg = RandomForestRegressor(n_estimators=100, oob_score=True, random_state=random_state)
svm_reg = SVR(kernel="linear", gamma='scale')
```

```python
for m in [lin_reg, tree_reg, forest_reg, svm_reg]:
    m.fit(X_train, y_train)
    burnout_predictions = m.predict(X_train)
    mse = mean_squared_error(y_train, burnout_predictions)
    rmse = np.sqrt(mse)
    r2 = m.score(X_train, y_train)

    print(f"{m}:\nRMSE = {rmse}\nr2 = {r2}\n")
```

Decision Tree Regressor and Random Forest Regressor have a suspiciously high R2 score measured on train dataset; Root-mean-square error (RMSE) of the Decision Tree Regressor is extremely small - overfitting? Cross-validation will show.

### Cross-validation

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
```

```python
def crossval(model, n_folds):
    scores = cross_val_score(model, X_train, y_train,
                                 scoring="neg_mean_squared_error", cv=n_folds)
    rmse_scores = np.sqrt(-scores)
    print(f"Cross-validation for {model}\nroot-mean-square error | R2: \n", end="")
    print(np.mean(rmse_scores), end=" | ")

    scores = cross_val_score(model, X_train, y_train,
                                 scoring="r2", cv=n_folds)
    print(np.mean(scores))
```

```python
crossval(lin_reg, 10)
```

```python
crossval(tree_reg, 10)
```

```python
crossval(forest_reg, 5) # Only five folds, otherwise it takes too long
```

```python
crossval(svm_reg, 5) # Only five folds, otherwise it takes too long
```

Decision Tree and Random Forest Regressors - now their R2s and root-mean-square errors look more plausible; the difference with the initially fitted models is huge which means that the Decision Tree model was clearly overfitted. As for Random Forest Regressor... it is not prone to overfitting, so here I am not sure yet why it performed so suspiciously good initially but luckily cross-validation does its job and returns more realistic scores. Nevertheless, Random Forest Regressor still shows the best R2 estimated on test dataset, followed by Linear Regression and SVR. 

I will use Grid Search, to find somewhat better combinations of parameters for most models.

#### Support Vector Regressor

```python
# try combinations of the following hyperparameters:
param_grid = [
    {'kernel':['rbf'], 'C':[1/10, 1, 10], 'epsilon': [0.025, 0.05]}
    ]

gridsearch(svm_reg, 3)
```

'C': 1, 'epsilon': 0.05, 'kernel': 'rbf'

After another evaluation, I am keeping 'kernel': 'rbf' and taking newly evaluated parameters 'C': 10 and 'epsilon': 0.05 as final / the most optimal for the Support Vector Regressor model.

#### Re-fitting models with the best found parameters

```python
lin_reg2 = LinearRegression()
tree_reg2 = DecisionTreeRegressor(max_depth=6, max_features=6, min_samples_leaf=25, random_state=random_state)
forest_reg2 = RandomForestRegressor(n_estimators=100, criterion='mse', max_features=4, oob_score=True, random_state=random_state)
svm_reg2 = SVR(kernel="rbf", gamma='scale', C=10, epsilon=0.05)
```

Decision Tree Regressor: much more realistic RMSE and R2 than with default parameters.<br>
Random Forest Regressor: again, too good RMSE and R2 but cross-validation showed that they get much closer to those from other models when applied to unseen data.<br>
Support Vector Regressor: slightly better RMSE and R2 than with default parameters.<br>

Support Vector Regressor has now a higher R2 than Linear Regressor.
